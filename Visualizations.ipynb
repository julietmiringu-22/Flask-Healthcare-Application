```python
import pandas as pd

df = pd.read_csv('C:/Users/Sharon/Desktop/True/flask data.csv', header=None)
df.columns = ["Age", "Gender", "Income", "Utilities", "Entertainment",
              "School Fees", "Shopping", "Healthcare"]

df.head()
```

```python
import matplotlib.pyplot as plt

# Convert Income column to numeric type
# The errors='coerce' parameter will convert non-numeric values to NaN
df["Income"] = pd.to_numeric(df["Income"], errors='coerce')

# Now we can calculate the mean
grouped = df.groupby("Age")["Income"].mean()

grouped.plot(kind='bar')
plt.title("Average Income by Age")
plt.xlabel("Age")
plt.ylabel("Income")

plt.savefig("age_income.png")
plt.show()
```

```python
df_melted = df.melt(id_vars=["Gender"], 
                    value_vars=["Utilities", "Entertainment", 
                                "School Fees", "Shopping", "Healthcare"],
                    var_name="Category",
                    value_name="Amount")

import seaborn as sns

sns.barplot(data=df_melted, x="Category", y="Amount", hue="Gender")

plt.title("Spending by Gender")
plt.xticks(rotation=45)

plt.savefig("gender_spending.png")
plt.show()
```
